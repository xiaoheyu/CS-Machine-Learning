```python
#%reset
import os

import json
import numpy as np
from copy import deepcopy
import pandas as pd
from collections import Counter as mset
```

```python
def tweets_k_means(K=25, initials="InitialSeeds.txt", tweetsFile="Tweets.json", output="tweets-k-means-output.txt"):
    os.remove(output)
# load json file to tweets vairiable
    tweets = []
    with open(tweetsFile) as f:
        for line in f:
            tweets.append(json.loads(line))

# Define the jaccard distance with string input A and B
    def jaccard_distance(A, B):
        intersection = len(sorted((mset(A.split()) & mset(B.split())).elements()))
        #print(sorted((mset(A.split()) & mset(B.split())).elements()))
        union = (len(A.split()) + len(B.split()))-intersection
        return float(1-intersection / union)

    
# SSE function
    SSE=0
    def sseCalculator(Center, tweets, cluster):

        for i in range(len(Center)):
            Classgroup = [tweets[m] for m in range(len(tweets)) if clusters[m] == i]
            for j in range(len(Classgroup)):
                #print(Center[i]["text"])
                #print(Classgroup[j]["text"])
                sserror = (jaccard_distance(Center[i]["text"], Classgroup[j]["text"]))**2

                #print(sserror)
                SSE += sserror
        return SSE
    
    
#initial lize centroids and convert Id to nemerical value
    C = []
    with open("InitialSeeds.txt") as f:
        for line in f:
            data = line.split()
            data = data[0].replace(",","")
            C.append(int(data))

# Convert controids from ID info to full tweets
    Ctweet = []
    for i in range(len(C)):
        for j in range(len(tweets)):
            if tweets[j]["id"] == C[i]:
                Ctweet.append(tweets[j].copy())
    #print(len(Ctweet))

    # Create old centroid for error calculation
    C_old = np.zeros(len(C))

    clusters = np.zeros(len(tweets))
    distances = np.zeros(len(Ctweet))
    error = 1
    errorlist=[]
    round = 1
    group=[]
    

    while error != 0:
        # Storing the old centroid values
        C_old = deepcopy(Ctweet)
        print("Round " + str(round) + " starts")
        # Create clusters matrix matching the original tweets data
        for i in range(len(tweets)):
            for j in range(len(Ctweet)):
                distances[j]= jaccard_distance(tweets[i]["text"], Ctweet[j]["text"])
            cluster = np.argmin(distances)
            clusters[i] = cluster

        # find all tweets for each class
        for i in range(len(Ctweet)):
            currentCumdist = 0
            smallestCumdist = 1000
            clusterTweets = [tweets[j] for j in range(len(tweets)) if clusters[j] == i]
            print(str(i) +" " + " th cluster has" + str(len(clusterTweets)) + " items")
            # find the tweet has the min distance to its class memebers and update center tweets in the class
            for m in range(len(clusterTweets)):
                for n in range(len(clusterTweets)):
                    currentCumdist += jaccard_distance(clusterTweets[m]["text"],clusterTweets[n]["text"])
                if currentCumdist < smallestCumdist:
                    smallestCumdist = currentCumdist
                    Ctweet[i] = clusterTweets[m]
                    print("the " +str(m) +"th tweet is selected as the center of class " + str(i))
                    currentCumdist = 0

        # detect whether center is still changing
        for i in range(len(Ctweet)):
            errorlist.append (jaccard_distance(Ctweet[i]["text"], C_old[i]["text"]))
        error = sum(errorlist)
        print("the error after round " +str(round)+  " is " + str(error) +"\n")
        errorlist=[]
        round +=1



# output SSE to txt
    group = []
    for i in range(len(Ctweet)):
        Classgroup = [tweets[m] for m in range(len(tweets)) if clusters[m] == i]
        group.append("class ID %s:" %int(i+1))
        for j in range(len(Classgroup)):
            group.append(Classgroup[j]["id"])
        with open(output, 'a') as f:
            print(group)
            for item in group:
                f.write("%s," %item)
            f.write("\n")
        group=[]
    with open(output,"a") as f:
        f.write("SSE: %s"%{SSE})
    


        
```
